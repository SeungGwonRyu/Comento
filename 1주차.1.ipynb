```python
def add(num1, num2):
    return num1 + num2

def minus(num1, num2):
    return num1 - num2

def multiply(num1, num2):
    return num1 * num2

def div(num1, num2):
    return num1 / num2

def divint(num1, num2):
    return num1 // num2

def calc(num1, num2, op) :
    if op == "+" : 
        return add(num1, num2)
    elif op == "-" : 
        return minus(num1, num2)
    elif op == '*' : 
        return multiply(num1,num2)
    elif op == '/' :
            return div(num1,num2)
    elif op == '//' :
            return divint(num1,num2)
```

```python
print(calc(2,3,'+'))
print(calc(2,3,'-'))
print(calc(2,3,'*'))
print(calc(2,3,'/'))
print(calc(2,3,'//'))
```

```python
# borad casting의 개념 : 산술 연산 중에 numpy가 다른 모양의 배열을 처리하는 방법을 설명한다. 이를 통해 서로 다른 크기의 배열간에도
# 연산을 수행할 수 있다. 코드를 간결하게 작성하고 효율적인 배열 연산을 가능하게 한다.
import numpy as np

# 같은 배열
a = np.array([1.0, 2.0, 3.0])
b = np.array([2.0, 2.0, 2.0])
result = a * b
print(result)

# 다른 배열 
c = np.array([1.0, 2.0, 3.0])
d = 2.0    # 같은 모양의 배열이 되도록 늘어남
result2 = c * d
print(result2)
```

```python
# map 자료형 : 키 - 값 쌍으로 이루어진 데이터를 저장하느 자료형이다. 각각 고유한 값이어야 하고 이는 효율적인 검색 연산을 제공하며, 데이터를 키 - 값 쌍으로 구성하여 유연한 자료구조를 제공한다.
# tuple 자료형 : 여러개의 요소로 이루어진 순서 있는 데이터 집합이다. 변경이 불가능한 자료형으로, 한번 생성된 후에는 내부요소를 추가, 수정, 삭제할 수 없다.

"""
 map자료형과 tuple자료형에 대해 조사하고 둘의 다른점이 무엇인지 간단히 설명 후 아래코드 작성
  => 문자열을 받고 각 알파벳이 몇개인지 출력하는 코드작성한다. 코드를 작성할때 아래 동작을 하는 함수를 구현후 꼭 사용해야 한다.
    @ 문자열을 받고 각 단어로 쪼갠후 이를 리스트에 저장해서 반환하는 함수
     * 예를들어 "abcde"를 받으면 ['a','b','c','d','e']를 반환한다.
    @알파뱃이 들어있는 리스트를 받으면 map자료형을 활용해 각 알파벳이 몇개인지 저장하고 이 map자료형을 반환하는 함수
     * 예를들어 ['a',b','a','d']를 받으면 [['a':2],['b':1],['d':1]]을 반환해야 한다.
    @ map 자료형을 출력해주는 함수
     * 예를 들어 [['a',2],['b',1],['d',1]]를 받으면 각 알파벳과 숫자를 짝지어 출력한다(출력형식은 자유)
"""
def toList(str):
    print("toList Called!!")
    print("   my ret is [{0}]",list(str))
    return list(str)

def makeDict(elem):
    print("makeDict Called!!")
    print("   my ret is [{0}]",[elem[0],1])
    return [elem[0],1]

def mapping(lst):
    print("mapping Called!!")
    print("   my lst is [{0}]",list(map(makeDict,lst)))
    return list(map(makeDict,lst))

def shuffle(mapped_result):
    shuffled_result = {}
    for key, val in mapped_result:
        shuffled_result[key] = shuffled_result.get(key, []) + [val]
        
    print("shuffle Called!!")
    print("   my ret is [{0}]",shuffled_result)
        
    return shuffled_result

def reduce(shuffled_data):
    ret = {}
    for key in shuffled_data.keys():
        ret[key] = sum(shuffled_data[key])
                
    print("reduce Called!!")
    print("   my ret is [{0}]",ret)
        
    return ret;

def repeatprint(reduced_data):
    var = ""
    
    for key,value in reduced_data.items():
        var += (key +  str(value))
        #try var += (key * value) 
    return var
```

```python
repeatprint(reduce(shuffle(mapping(toList("abcda")))))
```

```python
# 넘파이(Numpy):
# 넘파이는 파이썬의 수치 계산을 위한 핵심 라이브러리로, 다차원 배열(ndarray)을 처리하는 기능을 제공합니다. 넘파이 배열은 많은 양의 데이터를
# 효율적으로 저장하고 조작할 수 있으며, 수학적인 연산을 빠르고 효율적으로 수행할 수 있습니다. 넘파이는 선형 대수, 푸리에 변환, 난수 생성 등 다양한
# 수치 계산 작업에 사용됩니다.

# 판다스(Pandas):
# 판다스는 데이터 조작 및 분석을 위한 라이브러리로, 넘파이 기능을 기반으로 구축되었습니다. 판다스는 데이터프레임(DataFrame)이라는 자료구조를 제공하여
# 표 형식의 데이터를 다룰 수 있습니다. 데이터프레임은 열(Column)과 행(Row)으로 구성되며, 데이터를 쉽게 필터링, 정렬, 그룹화, 병합 등 다양한 작업을 
# 수행할 수 있습니다. 판다스는 데이터 처리 및 데이터 분석을 위한 강력한 도구입니다.

# 맷플롯립(Matplotlib):
# 맷플롯립은 데이터 시각화를 위한 파이썬 라이브러리로, 다양한 형식의 그래프와 플롯을 생성할 수 있습니다. 맷플롯립은 선 그래프, 산점도, 막대 그래프,
# 히스토그램, 파이 차트 등 다양한 시각화 방법을 제공합니다. 맷플롯립을 사용하여 데이터를 시각적으로 탐색하고 분석 결과를 보여줄 수 있습니다. 또한, 
# 맷플롯립은 넘파이나 판다스와 함께 사용될 수 있어 데이터를 그래프로 나타내는데 유용합니다.
```

```python
#Plotting categorical variables
import matplotlib.pyplot as plt
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
#Histograms
import matplotlib.pyplot as plt
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=True, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```
